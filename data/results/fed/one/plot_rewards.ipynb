```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
client_1 = pd.read_csv('client_1_rwd.csv')
client_2 = pd.read_csv('client_2_rwd.csv')
client_3 = pd.read_csv('client_3_rwd.csv')

#concatenate data
all_clients = pd.concat([client_1, client_2, client_3], axis=0)
```

```python
fig, ax = plt.subplots(figsize=(4,2))  # Use fig, ax for unpacking
sns.lineplot(data=all_clients, y='Value', x='Step', ax=ax)
ax.set_xlabel('Iterações (#)')
ax.set_ylabel('RME')
ax.set_ylim(350, 550)
ax.grid(True, linestyle='--')
```

```python
# Save the figure
fig.savefig('rewards.png', dpi=300, bbox_inches='tight')
```
